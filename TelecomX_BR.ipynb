```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Generate the cross-tabulation
tabela = pd.crosstab(df_limpo['account.PaymentMethod'], df_limpo['Churn'], normalize='index') * 100

# Blue shades for payment methods
cores_azul = ['#89fcb3', '#b1fcb3', '#d8fcb3', '#e9fccf', '#3366CC']

# Setup
indices = np.arange(len(tabela))
width = 0.35
fig, ax = plt.subplots(figsize=(10,6))

# Plot bars - blue for retained, red for churned
bars_retain = ax.bar(indices - width/2, tabela['No'], width,
                    color=cores_azul[:len(tabela)], label='Permaneceram')
bars_churn = ax.bar(indices + width/2, tabela['Yes'], width,
                   color='#D9534F', label='Abandonaram')

# Highlight the highest churn method
max_churn_idx = tabela['Yes'].idxmax()
max_churn_pos = list(tabela.index).index(max_churn_idx)
max_churn_val = tabela['Yes'].max()

# Add annotation box
ax.annotate('Métodos automáticos\nsão menos propensos\na evasão',
           xy=(max_churn_pos + width/2, max_churn_val),
           xytext=(20, 20), textcoords='offset points',
           bbox=dict(boxstyle='round,pad=0.5', fc='white', alpha=0.9),
           arrowprops=dict(arrowstyle='->'))

# Formatting
ax.set_xticks(indices)
ax.set_xticklabels(tabela.index, rotation=0, fontsize=10)
ax.set_yticks([])
ax.set_xlabel('Métodos de Pagamento', fontsize=12)
ax.set_ylabel('Percentagem', fontsize=12)
ax.set_title('Taxa de Retenção vs Evasão por Método de Pagamento', pad=20, fontsize=14)

# Custom legend
ax.legend(title='Status do Cliente', loc='upper center')

# Add value labels
def add_labels(bars):
    for bar in bars:
        height = bar.get_height()
        if height > 0:
            ax.text(bar.get_x() + bar.get_width()/2, height + 1,
                   f'{height:.1f}%', ha='center', va='bottom', fontsize=11)

add_labels(bars_retain)
add_labels(bars_churn)

# Clean up borders
for spine in ['left','top', 'right']:
    ax.spines[spine].set_visible(False)

plt.tight_layout()
plt.show()
```

```python
from scipy.stats import ttest_ind

grupo_no = df_limpo[df_limpo['Churn'] == 'No']['account.Charges.Total'].dropna()
grupo_yes = df_limpo[df_limpo['Churn'] == 'Yes']['account.Charges.Total'].dropna()

t_stat, p_val = ttest_ind(grupo_no, grupo_yes, equal_var=False)  # Welch t-test

print(f"T-statistic: {t_stat:.3f}")
print(f"P-value: {p_val:.4f}")

if p_val < 0.05:
    print("Diferença estatisticamente significativa entre os grupos")
else:
    print("Sem diferença estatisticamente significativa")
```

```python
import matplotlib.pyplot as plt
import numpy as np

contagem_contrato = df['account.Contract'].value_counts()
percentual_contrato = contagem_contrato / contagem_contrato.sum() * 100

cores_azul = ['#003366', '#336699', '#99CCFF']

def func(pct, allvals):
    absolute = int(pct/100.*sum(allvals))
    return f"{percentual_contrato.index[func.idx]}\n{pct:.1f}%"

func.idx = 0
def autolabel(pct):
    result = func(pct, percentual_contrato)
    func.idx += 1
    return result

fig, ax = plt.subplots(figsize=(7,7))
wedges, texts, autotexts = ax.pie(
    percentual_contrato,
    labels=None,
    autopct=autolabel,
    startangle=90,
    colors=cores_azul,
    textprops={'fontsize':12, 'color':'white'},
    labeldistance=0.5,
    pctdistance=0.5
)

ax.set_title('Percentual de Clientes por Tipo de Contrato')

# Anotação para Month-to-month com fundo azul claro translúcido
theta1, theta2 = wedges[0].theta1, wedges[0].theta2
ang = (theta1 + theta2) / 2
x = np.cos(np.deg2rad(ang))
y = np.sin(np.deg2rad(ang))

x_text = 1.3 * x
y_text = 1.3 * y - 0.3
ax.annotate(
    'Maioria com\n contrato\n mais propenço\n a evasão',
    xy=(x, y),
    xytext=(x_text, y_text),
    arrowprops=dict(facecolor='black', arrowstyle='->'),
    fontsize=11,
    ha='center',
    va='center',
    bbox=dict(boxstyle='round,pad=0.3', fc='#a8c6ff', alpha=0.7)
)


plt.show()
```

Contratos mensais concentram mais da metade da base, mas são responsáveis por quase metade do churn, confirmando alta volatilidade.

Investir em campanhas para conversão de cheque eletrônico para métodos automáticos, focando em facilitar a transição.

Introduzir incentivos financeiros imediatos para migração.

Recomendação:

Mapear e auditar casos atípicos para identificar causas raízes.

Ajustar processos de faturamento e atendimento para mitigar pontos de atrito que levam ao churn.

Contratos Longos
Criar ofertas exclusivas e diferenciadas para contratos anuais e bienais, aumentando valor percebido e barreiras de saída.

Implantar programa de fidelidade com benefícios tangíveis.

Pagamentos Automáticos
Automatizar lembretes e notificações para clientes com cheque eletrônico.

Oferecer incentivos financeiros para adoção imediata do débito automático.

Monitoramento Proativo
Implementar sistema de score para prever churn em clientes month-to-month.

Ações de retenção imediatas, como contato direto e ofertas personalizadas.
