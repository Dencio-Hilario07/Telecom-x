```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Generate the cross-tabulation
tabela = pd.crosstab(df_limpo['account.PaymentMethod'], df_limpo['Churn'], normalize='index') * 100

# Blue shades for payment methods
cores_azul = ['#89fcb3', '#b1fcb3', '#d8fcb3', '#e9fccf', '#3366CC']

# Setup
indices = np.arange(len(tabela))
width = 0.35
fig, ax = plt.subplots(figsize=(10,6))

# Plot bars - blue for retained, red for churned
bars_retain = ax.bar(indices - width/2, tabela['No'], width,
                    color=cores_azul[:len(tabela)], label='Permaneceram')
bars_churn = ax.bar(indices + width/2, tabela['Yes'], width,
                   color='#D9534F', label='Abandonaram')

# Highlight the highest churn method
max_churn_idx = tabela['Yes'].idxmax()
max_churn_pos = list(tabela.index).index(max_churn_idx)
max_churn_val = tabela['Yes'].max()

# Add annotation box
ax.annotate('M√©todos autom√°ticos\ns√£o menos propensos\na evas√£o',
           xy=(max_churn_pos + width/2, max_churn_val),
           xytext=(20, 20), textcoords='offset points',
           bbox=dict(boxstyle='round,pad=0.5', fc='white', alpha=0.9),
           arrowprops=dict(arrowstyle='->'))

# Formatting
ax.set_xticks(indices)
ax.set_xticklabels(tabela.index, rotation=0, fontsize=10)
ax.set_yticks([])
ax.set_xlabel('M√©todos de Pagamento', fontsize=12)
ax.set_ylabel('Percentagem', fontsize=12)
ax.set_title('Taxa de Reten√ß√£o vs Evas√£o por M√©todo de Pagamento', pad=20, fontsize=14)

# Custom legend
ax.legend(title='Status do Cliente', loc='upper center')

# Add value labels
def add_labels(bars):
    for bar in bars:
        height = bar.get_height()
        if height > 0:
            ax.text(bar.get_x() + bar.get_width()/2, height + 1,
                   f'{height:.1f}%', ha='center', va='bottom', fontsize=11)

add_labels(bars_retain)
add_labels(bars_churn)

# Clean up borders
for spine in ['left','top', 'right']:
    ax.spines[spine].set_visible(False)

plt.tight_layout()
plt.show()
```

```python
from scipy.stats import ttest_ind

grupo_no = df_limpo[df_limpo['Churn'] == 'No']['account.Charges.Total'].dropna()
grupo_yes = df_limpo[df_limpo['Churn'] == 'Yes']['account.Charges.Total'].dropna()

t_stat, p_val = ttest_ind(grupo_no, grupo_yes, equal_var=False)  # Welch t-test

print(f"T-statistic: {t_stat:.3f}")
print(f"P-value: {p_val:.4f}")

if p_val < 0.05:
    print("Diferen√ßa estatisticamente significativa entre os grupos")
else:
    print("Sem diferen√ßa estatisticamente significativa")
```
